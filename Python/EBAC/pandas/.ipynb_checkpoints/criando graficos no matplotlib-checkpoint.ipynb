```python
plt.plot(np.random.randn(50).cumsum(), 'k--')
#preenche o ultimo grafico com os paramtros
#o primeiro parametro e um array  com os dados, e o segundo um dado estetico
#onde cada letra tem e um parametro aperte shift+tab para vera s opcoes
# o k significa que e preto, mas poderia ser g, par verde, b para azul, r vermelhoc ciano, m magenta, y amarelo ou w opara branco
# o -- e para ser tracejado, mas tem muitas outras opçoes
#ou plt.plot?
```

```python
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=.3)
_= ax2.scatter(np.random.randn(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
#preenchenco os graficos aleatoriamente
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color = 'b', alpha=0.9)
    
#ajusta os espaçoes do grafico
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
plt.close('all')
```

```python
from numpy.random import randn
plt.plot(randn(30).cumsum(), 'go--')
```

```python
plt.close('all')
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
#linha preta
plt.plot(data, 'b-', drawstyle='steps-post', label='steps-post')
#linha azul
plt.legend(loc='best')
# legenda
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
#troca as nomeclaturas do eixo X
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['ano 1', 'ano 2', 'ano 3', 'ano 4', 'ano 5'],
                           rotation= 30, fontsize = 'small')
```

```python
#troca o titulo do grafico
ax.set_title('titulo do grafico')
#nomeia o eixo x do grafico
ax.set_xlabel('periodo')
```

```python
plt.close('all')
```

```python
fig = plt.figure(); ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),'k--', label='True')
ax.plot(np.random.randn(1000).cumsum(),'k.', label='three')
```

```python
ax.legend(loc='best')
```
