```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(14, 10))

plt.subplot(2, 3, 1)
sinasc['APGAR1'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuição APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Frequência')

plt.subplot(2, 3, 2)
sinasc['APGAR5'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribuição APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Frequência')

plt.subplot(2, 3, 3)
sinasc['SEXO'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Distribuição SEXO')
plt.xlabel('SEXO')
plt.ylabel('Frequência')

plt.subplot(2, 3, 4)
sinasc['GESTACAO'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Distribuição GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Frequência')

plt.subplot(2, 3, 5)
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', color='orange')
plt.title('Distribuição ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
media_apgar5_por_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean()

plt.figure(figsize=(6, 4))
media_apgar5_por_consultas.plot(kind='bar')

plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')

plt.xticks(rotation=0)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sinasc['SEXO'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Frequência de SEXO')
plt.xlabel('SEXO')
plt.ylabel('Frequência')

plt.subplot(2, 2, 2)
sinasc['GESTACAO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Frequência de GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Frequência')

plt.subplot(2, 2, 3)
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Frequência de ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Frequência')

plt.subplot(2, 2, 4)
sinasc['CONSULTAS'].value_counts().plot(kind='bar', color='orange')
plt.title('Frequência de CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
