```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc.boxplot(column='IDADEMAE')
plt.title('Boxplot da Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sinasc.boxplot(column='PESO')
plt.title('Boxplot do peso do bebê')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
sns.set(style="whitegrid")
sns.histplot(sinasc['IDADEMAE'], bins=30, kde=False)
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sns.set(style="whitegrid")
sns.histplot(sinasc['PESO'], bins=30, kde=False)
plt.title('Histograma do peso do bebê')
plt.xlabel('PESO')
plt.ylabel('Frequência')
plt.show()
```
